## Predict House Price Using Linear Regression

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyt
from sklearn import linear_model
%matplotlib inline
```

```python
#import data
df = pd.read_excel(r"home_price.xlsx")
df
```

```python
#Plot The data in matplotlib.pyplot using scatter plot
pyt.scatter(df.area,df.price, c="red", marker = "^")
pyt.xlabel("Area of the house (SFT)")
pyt.ylabel("Price of the house (US$)")
```

```python
#perform linear regression 
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
#plot the linera regression in pyplot and find the best fit line
pyt.scatter(df.area,df.price, c="red", marker = "^")
pyt.xlabel("Area of the house (SFT)")
pyt.ylabel("Price of the house (US$)")
pyt.plot(df.area,reg.predict(df[["area"]]))
```

```python
# use a data frame of areas of houses to predict their house prices and export a new csv 
#import csv file
df_n = pd.read_csv("areas.csv")
df_n.head()
```

```python
# predict the price using linear regression and create a new dataframe
price =reg.predict(df_n)
df_n["prices"] =price
df_n.head()
```

```python
# Export the data frame in a csv file
df_n.to_csv("prediction.csv", index=False)
```

# Find out the percapita income of canada in 2020 using linear regression

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyt
from sklearn import linear_model
%matplotlib inline
```

```python
#import data
df = pd.read_csv(r"D:\My Projects on programming\python\python-\ML\1_linear_reg\Exercise\canada_per_capita_income.csv")
df.head()
df.columns
df.rename(columns = {'per capita income (US$)':'per_capita_income_usd'}, inplace = True)
df.head()
```

```python
pyt.scatter(df.year,df.per_capita_income_usd, c="red", marker = "^")
pyt.xlabel("Year")
pyt.ylabel("Per capita income (US$)")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df.per_capita_income_usd)
```

```python
#plot the linera regression in pyplot and find the best fit line
pyt.scatter(df.year,df.per_capita_income_usd, c="red", marker = "^")
pyt.xlabel("Year")
pyt.ylabel("Per capita income (US$)")
pyt.plot(df.year,reg.predict(df[["year"]]))
```

```python
prediction = reg.predict([[2020]])
type(prediction)
p=prediction[0]
p
```

```python
print(f"the per capita income in 2020 is {p} usd")
```
